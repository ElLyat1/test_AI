```python
!pip install requests psycopg2-binary
import requests
import pandas as pd

books_response = requests.get("https://www.anapioficeandfire.com/api/books")
books_data = books_response.json()
books_df = pd.DataFrame(books_data)

print("Books DataFrame:")
print(books_df.head())

houses_response = requests.get("https://www.anapioficeandfire.com/api/houses")
houses_data = houses_response.json()
houses_df = pd.DataFrame(houses_data)

print("\nHouses DataFrame:")
print(houses_df.head())

# Фильтрация домов, у которых есть девиз
houses_with_words_df = houses_df[houses_df['words'].astype(bool)]

print("\nHouses with Words DataFrame:")
print(houses_with_words_df.head())
```
