class Chocolate: def **init**(self, price: float, add_on: str = None,
has_nuts: bool = False, num_nuts: int = 0): if price \<= 0: raise
ValueError(“Цена шоколадного батончика должна быть больше нуля”)
self.price = price self.add_on = add_on self.has_nuts = has_nuts
self.num_nuts = num_nuts

    def show_my_chocolate(self):
        if self.add_on:
            print(f"Шоколадный батончик, добавка: {self.add_on}")
        else:
            print("Обычный шоколадный батончик")

    def count_nuts(self):
        if self.has_nuts and self.num_nuts > 0:
            print('*' * self.num_nuts)
        else:
            print("Без орехов")

    def __str__(self):
        return f"Шоколадный батончик, {self.price} рублей"

# Демонстрация работы класса

choco1 = Chocolate(120.5, add_on=“Карамель”, has_nuts=True, num_nuts=5)
choco2 = Chocolate(90.0, has_nuts=False) choco3 = Chocolate(150.0,
add_on=“Мята”, has_nuts=True, num_nuts=8)

# Демонстрация метода show_my_chocolate

choco1.show_my_chocolate() \# Шоколадный батончик, добавка: Карамель
choco2.show_my_chocolate() \# Обычный шоколадный батончик

# Демонстрация метода count_nuts

choco1.count_nuts() \# \*\*\*\*\* choco2.count_nuts() \# Без орехов
